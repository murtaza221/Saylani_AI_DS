```python
students = {'name':["Ali","Faraz","Kamran","Faizan","Nasir","Asad","Danish"],
            'Age':[20,21,24,21,20,23,np.nan],
            'Python':[20,40,78,96,100,78, 80],
            'Numpy':[78,87,np.nan,89,75,95,85],
            'Pandas':[40,75,85,48,60,89, 96]}

data = pd.DataFrame(students)
data
```

```python
data.fillna(0)
```

```python
data.fillna({'Age':data['Age'].mean(),
             "Python":0,"Numpy":0,"Pandas":0},inplace=True)
```

```python
data
```

```python
data ['totalmarks']=300
data['obtanMark'] = data.Python+data.Numpy+data.Pandas
data
```

```python
data['Result']=["Pass" if m>150 else "Fail" for m in data["obtanMark"]]
data
```

```python
data['percentage']= (data['obtanMark']/data['totalmarks'])*100
data
```

```python
data['Grade']= ["A+" if perecent>=80 
                else "A" if perecent>=70
                else "B" if perecent>=60
                else "C" if perecent>=50
                else "Fail"
               
                for perecent in data ['percentage']
                ]
```
