### Modification
    add, delete, append/replace, extend

```
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)
motorcycles[0] = 'ducati'
print("Updated: ",motorcycles)
motorcycles.append('honda')
print("Element Added",motorcycles)
```

```
#Extend list
country= "America"
# string is iterable and immutable
motorcycles.extend(country)
motorcycles
```

```
num=[1,3,4,56445,34,34,23,245,4565,6767,5, 'A', 'm', 'e', 'r', 'i', 'c', 'a']
num.index("A",3)
print("A is found at ",num.index("A",3))
```

```
num1=[1,3,4,56445,34,34,23,245,4565,6767,5]
num1
```

```
#remove
num1.remove(6767)
print("Elemnent removed\n", num1)
```

```
#reverse
num1.reverse()
num1
```

```
#sort
num1.sort(reverse = True)
print(num1)
num1.sort()
print(num1)
```

```
#pop
popped_element = num1.pop()
print("Element popped:",popped_element)
print(num1)
```

```
#sort vs sorted
num1=[1,3,4,56445,34,34,23,245,4565,6767,5]
print("List:   ",num1)

num2=num1
num2.sort()
print("Sort:   ",num2)

num1_sorted =sorted(num1)
print("Sorted: ",num1_sorted)
#sorted is the function of Python
```
