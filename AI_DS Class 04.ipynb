# List comprehension

for loop when using if only:
    #[<expression> <forloopbody> <if statement>]

for loop when using if only:
    #[<body_of_if> <if_statement> <else_statement_with_body> <forloopbody> ]

```
squares=[]
for a in range(10):
  square =a**2
  squares.append(square)
squares
```

```
# above code through list comprehension
hello = [a**2 for a in range(10)]
print(hello)
```

```
[a*2 for a in range(10)]
```

```
[a/2 for a in range(10)]
```

```
[a for a in range(100) if a%5==0 ]
```

```
["Div by 5" if a%5==0  else "Not div by 5" for a in range(50) ]
```

```
["Odd" if a%2!=0  else "Even" for a in range(10) ]
```

```
[f"Odd => {(a)}" if a%2!=0  else f"Even => {(a)}" for a in range(10) ]
```
