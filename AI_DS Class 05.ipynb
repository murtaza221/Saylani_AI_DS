### Dictionary
    Dictionary have no index it has keys.

```
dictionary={'name':'Sami','age':23,'qual':'BS'}
dictionary
```

```
dictionary1={'name':'Sami','age':23,'qual':'BS','Exp':5}
dictionary['name']
```

```
#to update
dictionary['name']="Iftikhar"

dictionary['name']
dictionary
```

```
#printing only keys
for key in dictionary.keys():
  print (key)
```

```
#printing keys with values using keys
for key in dictionary.keys():
  print (dictionary[key])
```

```
#printing only values
for value in dictionary.values():
  print (value)
```

```
#printing both value and key at the same time
for k,v in dictionary.items():
  print(k,"\t:",v)
```

```
#printing both value and key at the same time with sequence number
for k,v in enumerate(dictionary.items()):
  print(k,"\t:",v)
```

```
#printing both value and key at the same time with sequence number
for a,(b,c) in enumerate(dictionary.items()):
  print(a,b,c)
```

```
#Dictionary methods/functions
#item key pop popitem setdefault values
dictionary1={'name':'Sami','age':23,'qual':'BS','Exp':5}
print(dictionary1.pop('name')) # we have to give key in pop
print(dictionary1.popitem()) # pop the last entered item
print(dictionary1.setdefault(''))
```
