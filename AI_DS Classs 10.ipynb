## Python Generators

##### in function it is <b>return</b>
##### in generator it is <b>yield</b>
to study python decorators


```
#demo code
def square_numbers(num):
  result = []
  for i in num:
    result.append(i*i)
  return result
```

```
#driver code 
my_nums= square_numbers([1,2,3,4,5,6])

print(my_nums)
```

```
# Doing same task with generators
def square_numbers_generators(nums):
  for i in nums:
    yield(i*i)
```

```
my_num = square_numbers_generators([6,7,8,9,10])
print(my_num)

for inum in  my_num:
  print(inum)
```
