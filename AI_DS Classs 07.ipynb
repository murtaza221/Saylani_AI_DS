```
def func_num():
  alist=[]
  for x in range(5):  
    x=int(input("Enter num"))
    alist.append(x)

  return max((alist))

func_num()
```

```
def func_oddeven():
  x=int(input("Enter num"))
  if(x%2==0): 
    print(x," is Even number")
  else:
    print(x," is odd number")

func_oddeven()
```

```
def func_compare():
  myval=22
  x=int(input("Enter your value"))
  if(x==myval): 

    print("your value is same as myvalue")
  elif(x>myval):
    print("your value is greater than myvalue")
  else:
    print("your value is less than myvalue")
func_compare()
```
